```python
spark.sql("""
CREATE OR REPLACE TABLE dim_customer_type2 (
    customer_id INT,
    customer_name STRING,
    city STRING,
    effective_date DATE,
    end_date DATE,
    is_current BOOLEAN
)
""")
```

```python
spark.sql("""
INSERT INTO dim_customer_type2 VALUES
(101,'Neha','Pune','2026-01-01',NULL,TRUE),
(102,'Rahul','Mumbai','2026-01-01',NULL,TRUE),
(103,'Amit','Delhi','2026-01-01',NULL,TRUE)
""")
```

```python
display(spark.table("dim_customer_type2"))
```

```python
updates = [
    (101, "Neha", "Hyderabad"),
    (104, "Priya", "Pune")
]

updates_df = spark.createDataFrame(
    updates,
    ["customer_id", "customer_name", "city"]
)

updates_df.createOrReplaceTempView("customer_updates")
```

```python
updates = [
    (101, "Neha", "Hyderabad"),
    (104, "Priya", "Pune")
]

updates_df = spark.createDataFrame(
    updates,
    ["customer_id", "customer_name", "city"]
)

updates_df.createOrReplaceTempView("customer_updates")
```

```python
spark.sql("""
UPDATE dim_customer_type2
SET
    end_date = current_date(),
    is_current = false
WHERE customer_id IN (
    SELECT customer_id
    FROM customer_updates
)
AND is_current = true
""")
```

```python
spark.sql("""
INSERT INTO dim_customer_type2
SELECT
    customer_id,
    customer_name,
    city,
    current_date(),
    NULL,
    true
FROM customer_updates
""")
```

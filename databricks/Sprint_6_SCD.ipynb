```python
customers = [

    (101, "Neha", "Pune"),

    (102, "Rahul", "Mumbai"),

    (103, "Amit", "Delhi")

]

columns = [

    "customer_id",

    "customer_name",

    "city"

]

df = spark.createDataFrame(customers, columns)

df.write \
    .mode("overwrite") \
    .format("delta") \
    .saveAsTable("dim_customer_type1")

display(spark.table("dim_customer_type1"))
```

```python
updates = [

    (101, "Neha", "Hyderabad"),

    (104, "Priya", "Pune")

]

updates_df = spark.createDataFrame(
    updates,
    columns
)

updates_df.createOrReplaceTempView(
    "customer_updates"
)
```

```python
spark.sql("""

MERGE INTO dim_customer_type1 target

USING customer_updates source

ON target.customer_id = source.customer_id

WHEN MATCHED THEN

UPDATE SET *

WHEN NOT MATCHED THEN

INSERT *

""")
```
